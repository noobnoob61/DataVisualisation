```python
df=pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
#Relplot -plotting with relplot
#         -gives scatter graph as default
      
sns.relplot(x='reading score',y='writing score',data=df,hue='gender',col='test preparation course')


#reading and writing scores have positive relation as according to the graph
#which tells that the better the reading skills the better the writing skills
#candidates not have taken any preparation courses are more then those with courses
```

```python
sns.relplot(x='reading score',y='writing score',data=df,hue='gender',col='test preparation course',kind='line')

#relplot as a line graph
```

```python
#catplot -plotting using catplot
#        - contains both catbolical value and numerical values
sns.catplot(x='gender',y='math score',data=df,kind='violin')
```

```python
#plotting with barplot

sns.barplot(x='gender',y='writing score',data=df,hue='test preparation course')

#candidates taken preparation course relatively scored higher then those of not taken
#females scored more in writing test
```

```python
sns.catplot(x='gender',y='writing score',data=df,hue='test preparation course',kind='bar')
plt.title('Student performance')
plt.ylabel('writing scores')
plt.xlabel('gender')
plt.show()
```

```python
df['gender'].value_counts()    #counts number of male and females
key=list(df['gender'].value_counts().keys()) #stores keys
value=list(df['gender'].value_counts().values) #store values

plt.title('Student preparation',loc='left') #placinf title on left

ls=[0.0,0.2]
plt.pie(value,labels=key,wedgeprops={'ec':'k'},explode=ls,autopct="%1.2f%%",shadow=True) #piechart

sns.set_theme(rc={'figure.figsize':(6,6)}) #increasing size
plt.savefig("Student data") #saving the graph
plt.show() #displaying the graph
```
